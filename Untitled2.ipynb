```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np

# Load and clean data
df = pd.read_csv("Quote-Equity-RELIANCE-EQ-09-03-2025-to-09-04-2025.csv")

# Clean column names
df.columns = df.columns.str.strip().str.upper().str.replace(" ", "_")

# Parse numeric values (remove commas)
cols_to_clean = ['OPEN', 'HIGH', 'LOW', 'CLOSE']
for col in cols_to_clean:
    df[col] = df[col].str.replace(',', '').astype(float)

# Optional: sort by date
df['DATE'] = pd.to_datetime(df['DATE'], format="%d-%b-%Y")
df = df.sort_values('DATE')

# Optional: keep last N days
df = df[-30:]

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))
candlestick_width = 0.6
gap = 1

for i, row in df.iterrows():
    o, h, l, c = row['OPEN'], row['HIGH'], row['LOW'], row['CLOSE']
    color = 'green' if c >= o else 'red'
    x = i * (candlestick_width + gap)
    ax.plot([x, x], [l, h], color=color)
    rect_y = o if c <= c else c
    height = abs(c - o)
    ax.add_patch(Rectangle((x - candlestick_width / 2, min(o, c)), candlestick_width, height, color=color))

# X-axis formatting
ax.set_title("Candlestick Chart for RELIANCE")
ax.set_xlabel("Date")
ax.set_ylabel("Price")

xticks = np.arange(0, len(df) * (candlestick_width + gap), step=(candlestick_width + gap) * 5)
xlabels = df['DATE'].dt.strftime('%d-%b')[::5]
ax.set_xticks(xticks)
ax.set_xticklabels(xlabels, rotation=45)

plt.tight_layout()
plt.show()
```
