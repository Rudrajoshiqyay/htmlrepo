```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np
import io
import base64
from datetime import datetime

def handler(request):
    # Load and clean data
    df = pd.read_csv("Quote-Equity-RELIANCE-EQ-09-03-2025-to-09-04-2025.csv")
    df.columns = df.columns.str.strip().str.upper().str.replace(" ", "_")

    cols_to_clean = ['OPEN', 'HIGH', 'LOW', 'CLOSE']
    for col in cols_to_clean:
        df[col] = df[col].str.replace(',', '').astype(float)

    df['DATE'] = pd.to_datetime(df['DATE'], format="%d-%b-%Y")
    df = df.sort_values('DATE')
    df = df[-30:]

    # Create candlestick chart
    fig, ax = plt.subplots(figsize=(12, 6))
    candlestick_width = 0.6
    gap = 1

    for i, row in enumerate(df.itertuples()):
        o, h, l, c = row.OPEN, row.HIGH, row.LOW, row.CLOSE
        color = 'green' if c >= o else 'red'
        x = i * (candlestick_width + gap)
        ax.plot([x, x], [l, h], color=color)
        ax.add_patch(Rectangle((x - candlestick_width / 2, min(o, c)), candlestick_width, abs(c - o), color=color))

    ax.set_title("Candlestick Chart for RELIANCE")
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")

    xticks = np.arange(0, len(df) * (candlestick_width + gap), step=(candlestick_width + gap) * 5)
    xlabels = df['DATE'].dt.strftime('%d-%b')[::5]
    ax.set_xticks(xticks)
    ax.set_xticklabels(xlabels, rotation=45)

    plt.tight_layout()

    # Save plot to a bytes buffer
    buffer = io.BytesIO()
    plt.savefig(buffer, format='png')
    buffer.seek(0)
    image_base64 = base64.b64encode(buffer.read()).decode('utf-8')
    buffer.close()

    # Return HTML with embedded image
    html = f"""
    <html>
    <head><title>Candlestick Chart</title></head>
    <body>
        <h2>Candlestick Chart for RELIANCE</h2>
        <img src="data:image/png;base64,{image_base64}" alt="Candlestick Chart"/>
    </body>
    </html>
    """

    return {
        "statusCode": 200,
        "headers": {"Content-Type": "text/html"},
        "body": html
    }

```
